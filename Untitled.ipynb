```python
from time import perf_counter
def timer(f):
    def _f(*arg):
        t0 =perf_counter()
        f(*arg)
        return perf_counter()-t0
    return _f
```

```python
class Node():
    def __init__(self,value,next):
        self.value = value
        self.next = next
    def __str__(self):
        return str(self.value)
    def __repr__(self):
        return 'Node ('+ str(self.value) + ')' 
```

```python
class LinkList():
    def __init__(self,lst=None):
        self.next = None
        if lst !=None:
            for val in lst:
                self.add(val)
    def __str__(self):
        p = self.next
        s='->'
        while p!=None:
            s+=str(p.value)
            s+='->'
            p = p.next
        s+='None'
        return s
    def add(self,value):
        self.next = Node(value,self.next)
    def Insert(self,value,postion):
        n1 = Node(value,None)
        p1 = self
        i=1
        while p1!= None and i!=postion:
            p1 = p1.next
            i+=1
        if p1!=None and i==postion:
            n1.next = p1.next
            p1.next =n1
    def remove(self,val):
        p2 = self
        while p2.next != None:
           
            p3 = p2.next
            print(p3.value)
            if p3.value== val:
                p2.next = p3.next
                return
            p2 =p3
```

```python
def json_handle(json_obj,attribute):
    dialy_json = {}
    for arr in attribute:
        dialy_json[arr] = []
    tmp = json_obj['daily']['data']
    for t in tmp:
        for arr in attribute:
            dialy_json[arr] = t[arr]
    return dialy_json
def remote_weather():
    weatherdata = requests.get('https://api.darksky.net/forecast/e5bed245988f142474927ad3b4a39910/37.8267,-122.4233')
    handleddata = json.loads(weatherdata.text)
    return handleddata
def neigborhood_weather():
    with open('sample.json',encoding='utf-8') as file:
        jsoncontent = json.load(file)
    return jsoncontent

def func_weather(remote = False):
    if remote:
        jsondata = remote_weather()
    else:
        jsondata = neigborhood_weather()
    attribute = ['temperatureMin','temperatureMax','sunsetTime']
    tm = json_handle(jsondata,attribute)
    for k, v in tm.items():
        print(k,v)
```

```python
import sklearn.datasets as datasets
import pandas as pd
iris = datasets.load_iris()
df =pd.DataFrame(iris.data,columns = iris.feature_names)
y= iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
